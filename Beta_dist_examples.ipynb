```python

#------------------------------------------------------------
# Define the distribution parameters to be plotted
alpha_values = [0.5, 1.5, 3.0, 0.5]
beta_values = [0.5, 1.5, 3.0, 1.5]
linestyles = ['-', '--', ':', '-.']
x = np.linspace(0, 1, 1002)[1:-1]

```

```python

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(10, 7.5))

for a, b, ls in zip(alpha_values, beta_values, linestyles):
    dist = beta(a, b)

    ax.plot(x, dist.pdf(x), ls=ls, c='black',
             label=r'$\alpha=%.1f,\ \beta=%.1f$' % (a, b))

plt.xlim(0, 1)
plt.ylim(0, 3)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\alpha,\beta)$')
plt.title('Beta Distribution')

plt.legend(loc=0)
plt.show()
```

```python

#------------------------------------------------------------
# Define the distribution parameters to be plotted
alpha_values = [10.2, 10.3]
beta_values = [18.8, 16.2]
linecols = ['r', 'b']
x = np.linspace(0, 1, 1002)[1:-1]
```

```python
#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(6, 4.5))

for a, b, lc in zip(alpha_values, beta_values, linecols):
    dist = beta(a, b)

    ax.plot(x, dist.pdf(x), c=lc,
             label=r'$\alpha=%.1f,\ \beta=%.1f$' % (a, b))
    plt.axvline(x=0.3, c='r', ls=':')
    plt.axvline(x=0.35, c='b', ls=':' )

plt.xlim(0, 1)
plt.ylim(0, 5.1)
plt.title ('Example (a)', fontsize=18)
plt.xlabel('Likelihood of possessing contraband', fontsize=16)
plt.ylabel(r'Density $p(x|\alpha,\beta)$', fontsize=16)
#plt.title('Beta Distribution')

plt.legend(loc=0, fontsize=14)
plt.show()
```

```python
ref = 'https://raw.githubusercontent.com/jkastelan/DAA2018_jlk635/master/Data/TXstopsPivot.csv'

thresholds = pd.read_csv(ref)

```

```python
thresholds.head()
```

```python

thresholds['Group'] = ' '
thresholds = thresholds[['Group', 'Asian_Thresh', 'Black_Thresh', 'Hispanic_Thresh', 'White_Thresh']]

renameColumns = {'Asian_Thresh':'Asian', 
                 'Black_Thresh':'Black', 
                 'Hispanic_Thresh':'Hispanic', 
                 'White_Thresh':'White'}

thresholds.rename(columns = renameColumns, inplace = True)
```

```python
thresholds.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
plt.figure(figsize=(8,2))

sns.set_style("whitegrid")
dataSubset = thresholds[['Asian','Black','Hispanic','White']]

sns.boxplot(data=dataSubset,orient="h",palette="Set3") 
plt.title('Comparison of search thresholds', fontsize = 18)
plt.ylabel('Race', fontsize = 16)
plt.xlabel('Threshold value', fontsize = 16)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 15);
```

$\mathrm{Search\ rate} = \frac{\mathrm{ Number\ of\ searches\ (per\ county,\ per\ racial\ group)}}{\mathrm{Number\ of\ stops\ (per\ county,\ per\ racial\ group)}}$

$\mathrm{Hit\ rate} = \frac{\mathrm{ Number\ of\ hits\ (per\ county,\ per\ racial\ group)}}{\mathrm{Number\ of\ searches\ (per\ county,\ per\ racial\ group)}}$
